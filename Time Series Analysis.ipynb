# ARIMA

Traditional regression models often fail to capture time series dynamics. However, by incorporating correlations over time, utilizing autoregressive (AR), integrating nonstationary components, and employing autoregressive moving average (MA) models, this problem can be addressed. These enhancements improve predictions, resulting in the widely-used autoregressive integrated moving average (ARIMA) model for time series forecasting (Shumway and Stoffer, 2017, p.75; www.statsmodels.org, n.d.)

The time series analyzed here is from Coursera, comprising 752 entries spanning from 2021-03-31 to 2024-03-26. However, it is expected that there was an increase in these stocks during the pandemic period, which was an atypical period. Therefore, I will sort the data to exclude this period, aiming to obtain a more reliable behavior that could make sense for forecasting with precision.

```python
cour_df = pd.read_csv("COUR.csv")
cour_df
```

```python
cour_df.info()
```

I will convert the 'Date' column to datetime format and set it as the index using the 'to_datetime()' and 'set_index()' functions because it can be more efficient when retrieving data. Additionally, I will use the 'sort_index()' method to ensure that the index is in the correct chronological order. 

I am using the '.asfreq()' method with the letter 'B' to represent a Monday-to-Friday frequency for this data. I'm doing this to address the warning message I encountered while using the model: 'ValueWarning: A date index has been provided, but it has no associated frequency information and so will be ignored when, for example, forecasting'.

```python
cour_df['Date'] = pd.to_datetime(cour_df['Date'])
cour_df.set_index('Date', inplace=True)
cour_df = cour_df.asfreq('B')
cour_df = cour_df.sort_index()
```

Before removing a portion of the time series, I inspect if it will be necessary using the plots created by the 'plotds' function in class. By examining these plots, I can gain insights into the behavior of the time series, identify any patterns or trends, and determine if any further preprocessing or analysis is necessary.

The 'plotds' function provides the time series chart for the 'Close' column, as well as the autocorrelation function (ACF) and the partial autocorrelation function (PACF) that can provide insights regarding to the tendency, seasonality and anomalies. I am using 'nlag = 50' to retrieve the last 50 values in the time series, which is usualy a reasonable approach for examining recent data points and their relationships.

```python
def plotds(xt, nlag=50, fig_size=(12, 10)):
    if not isinstance(xt, pd.Series):
        xt = pd.Series(xt)
    plt.figure(figsize=fig_size)
    layout = (2, 2)
    ax_xt = plt.subplot2grid(layout, (0, 0), colspan=2)
    ax_acf = plt.subplot2grid(layout, (1, 0))
    ax_pacf = plt.subplot2grid(layout, (1, 1))
    xt.plot(ax=ax_xt)
    ax_xt.set_title('Time Series: COUR')
    plot_acf(xt, lags=50, ax=ax_acf)
    plot_pacf(xt, lags=50, ax=ax_pacf)
    plt.tight_layout()
    plt.show()

plotds(cour_df['Close'])
```

The time series plot showed a trend of reduction as the pandemic was coming to an end (around January 2022); subsequently, it tended to stabilize with small fluctuations of increase and decrease.

Considering the autocorrelation function (ACF) and partial autocorrelation function (PACF) up to 50 lags, I found a considerable high positive correlation (1.00 to ~0.75) of correlation coefficient for ACF. This pattern does not show seasonality, only a decreasing tendency in the stock prices, indicating that the time series is not stationary.

For PACF, which indicates the partial correlation between the current value of the time series and its value at a particular lag after removing the effects of intervening observations, I found high correlation only in the first two lags. For the other lags, the linear correlation between the current value of the time series and its value at the specified lag is too weak (~0.10) and did not show seasonality.

Therefore, the pandemic period appeared to be an anomaly and may lead to unreliable forecasting. Because of this, I will remove the data from 2021-03-31 to 2022-01-01, which corresponds to the pandemic perio..

Here, I filtered the data to commence from 2022-01-01, which could mark the end of the pandemic period. Now the dataset is called 'cour_filt' and compreends the period of 2022-01-03 to 2024-03-26.

```python
start_date = '2022-01-01'
cour_filt = cour_df[cour_df.index >= start_date]
cour_filt
```

I will apply inspect again the time series, ACF and PACF.

```python
def plotds(xt, nlag=50, fig_size=(12, 10)):
    if not isinstance(xt, pd.Series):
        xt = pd.Series(xt)
    plt.figure(figsize=fig_size)
    layout = (2, 2)
    ax_xt = plt.subplot2grid(layout, (0, 0), colspan=2)
    ax_acf = plt.subplot2grid(layout, (1, 0))
    ax_pacf = plt.subplot2grid(layout, (1, 1))
    xt.plot(ax=ax_xt)
    ax_xt.set_title('Time Series: COUR')
    plot_acf(xt, lags=50, ax=ax_acf)
    plot_pacf(xt, lags=50, ax=ax_pacf)
    plt.tight_layout()
    plt.show()

plotds(cour_filt['Close'])
```

As we can see in the chart above, now excluding the pandemic period, the time series seems to have high fluctuations with two peaks, one in January 2022 and another in December 2023. In addition, the ACF plot does not seem to present significant peaks at the lag interval that could correspond to the seasonality period. 


Taking up to 50 lags, the ACF and PACF showed similar results when the time series had the pandemic period; however, the ACF showed the correlation coefficient decreasing from 1.00 to 0.50. This means that the correlation coefficient experienced a reduction from 0.75 to 0.50. Overall, the slow decay in the ACF suggest that the time series is not stationary.

### Step 1: Check for Stationarity

This is step is important because autoregressive models, such ARIMA are similar to linear regression, being essential to remove any correlation and collinearity with the past making the time series stationary to reduce bias (RAVAL, 2024).

At first glance, we can see that the data does not appear to be stationary due to the fluctuations that typically result in different means and variances. However, to certify this, I will use the Dickey-Fuller test.  

```python
result = adfuller(cour_filt['Close'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

Since the p-value is less than the significance level of 0.05, I would reject the null hypothesis, suggesting that the data is stationary. However, as this value is too close, I will apply first-order differencing to ensure a more robust result.

Below, I am using the Quantile-Quantile plot, or Q-Q plot, to visually assess whether the distribution of the 'Close' price follows a normal distribution by comparing it with a theoretical normal distribution. This is important because time series models often assume a normal distribution. As we can see below, the data closely resembles a normal distribution.

```python
x = sm.qqplot(cour_filt['Close'], line = 's')
plt.show()
```

### Step 2: Differencing

##### First-order differencing

I will use first-order differencing to render the data stationary and enable more reliable forecasting. This involves employing the 'diff()' function with an argument of 1 to compute the difference between each element and its immediately preceding element. Subsequently, I will use the 'dropna()' function to eliminate any resulting 'NaN' values.

```python
first_order_diff = cour_filt['Close'].diff(1).dropna()
first_order_diff.head()
```

Below is a comparative chart between the original time series and the time series after applying first-order differencing, rendering the data clearly stationary. Then, I will apply the Dickey-Fuller test to verify this.

```python
fig, ax = plt.subplots(2, sharex = True)
fig.set_size_inches(5.5, 5.5)
cour_filt['Close'].plot(ax = ax[0], color = 'b')
ax[0].set_title('Close values of COUR (Jan 2022 - Mar 2024)')
first_order_diff.plot(ax = ax[1], color = 'r')
ax[1].set_title('First-order differences of COUR (Jan 2022 - Mar 2024)')
plt.show()
```

Below, we observe that the Dickey-Fuller test yielded the most reliable p-value. Consequently, I can now confirm that the time series is stationary and I will use the first-order differencing in the ARIMA model. 

```python
adf_result = adfuller(first_order_diff)

print('ADF Statistic: %f' % adf_result[0])
print('p-value: %f' % adf_result[1])
```

```python
plotds(first_order_diff, nlag = 50)
plt.show()
```

These results, where the ACF and PACF both cut off after the first lag, suggest that there is a significant correlation at the lag 1, but the correlations at subsequent lags quickly drop near to zero.  

To determine the optimal component values, I will assess the Akaike Information Criterion (AIC) value before fitting the model.

To perform the ARIMA model I need three components: (p, d, q)

    p = autoregressive (AR) component.

    d = differencing component.

    q = moving average (MA) component.

### Step 3: Train and test split

I will split the 'cour_filt['Close']' time series into 80% train, 20% test before perform the ARIMA model. This is important because we can use the training data for model selection to ensure that the model parameters will be chosen based on the available information without leaking information from the test set, thus preserving against overfitting. In addition, I will used the code from line 3 to ensure that the initial portion of the data is used for the training set, while the remaining portion comprises the test set, ensuring that the past data will be used to train the model, and future data will be used to test it (Brownlee, 2023).

```python
X = cour_filt['Close'].values
size = int(len(X) * 0.80)
train_data, test_data = X[0:size], X[size:len(X)]
train_ar = [x for x in train_data]                      
predictions = list()

train_data.shape, test_data.shape
```

### Step 4: Select AR and MA components

#### Using ACF and PACF plots to find (p, d, q)

Considering these plots, we can see that both indicate the first lag as the most significant, suggesting that we can consider p = 1 and q = 1 as well. Thus, I used these (p=1, d=1, q=1) components to fit the ARIMA model.
I will used the training set to fit the ARIMA model to avoid overfitting 

I am using the parameter 'missing='drop'' to remove any observation with NaN (www.statsmodels.org, n.d.)..

```python
arima_model = ARIMA(train_data, order=(1,1,1), missing='drop')
model = arima_model.fit()
print(model.summary())
```

The summary shows that the AIC value, which measures the goodness of fit, is equal to 740.140. However, more comparisons need to be made in order to improve the model's performance.

The ar.L1 coefficient represents the autoregressive (AR) term at lag 1, indicating how much the current value of the data is influenced by the previous value. In this case, I foundry weak positive autocorrelation at lag 1, suggestingry small influence.

The ma.L1 coefficient represents the moving average (MA) term at lag 1, indicating how much the current value is influenced by the error from the previous prediction. It suggestreasonablerong negative correlation with the previous error term. That is, if the previous prediction error goes up, the current value is expected to go down0b23109946 units.

The sigma2 is the variance of the white noise term, indicating the amount of random fluctuation or noise in the data that is not accounted for by the AR and MA terms.

####  Using AIC value to find (p, d, q)

Here I am using the Akaike Information Criterion (AIC) to evaluate and determine the most suitable components for the ARIMA model.

I will use the nested loop below to calculate the AIC value in the training set for the same reason as exposed above, that is to preserve leaking information from the test set which helps to prevent overfitting. 

```python
aicVal=[]
for d in range(1,3):
    for ari in range(0, 3):
        for maj in range(0,3):
            try:
                arima_obj = ARIMA(train_data, order=(ari,d,maj))
                arima_obj_fit=arima_obj.fit()
                aicVal.append([ari, d, maj, arima_obj_fit.aic])
            except ValueError:
                pass
print(aicVal)
```

The smallest AIC result indicates the best model performance; in my case, the smallest AIC was [0, 1, 0, 736.411].

As we can see, the results suggest first-order differencing, however the components autoregressive (AR) and moving average (MA) are not considered.

The absence of AR and MA terms means that the current value of the time series not depend on its previous values neighter the previous errors terms.

####  Using auto_arima function to find (p, d, q)

Here, I will use another option to find the optimal values for the components p, d, and q in the ARIMA model by applying the 'auto_arima' function (alkaline-ml.com, 2024).

```python
model_auto_arima = pm.auto_arima(train_data, start_p=0, start_q=0, test='adf', max_p=3, max_q=3, d=None, trace=True, missing='drop')

print(model_auto_arima.summary())
```

The use of the 'auto_arima' function resulted in an identical result to what I found in the nested loop above, corroborating that the best components of data I can use with the ARIMA model in my case are (p, d, q) = (0, 1, 0).

### Step 5: Build the model

I will fit the ARIMA model with the training data, using the order (p=0, d=1, q=0), as it showed better results in the previous step.

```python
model_td = ARIMA(train_data, order=(0,1,0), missing='drop')
model_td_fit = model_td.fit()

model_td_fit.summary()
```

#### Model's residuals

Considering the model fitting with the training set above I will use the '.plotdiagnosis' to examine the residuals of this model.

```python
model_td_fit.plot_diagnostics(figsize = (12, 10))
plt.suptitle('Residual Diagnostic: COUR', fontsize=16, fontweight='bold')
plt.show()
```

The plot in the **top left** shows that the residual errors vary uniformly around a mean of zero.

In the **top right**, a density histogram is presented, suggesting a normal distribution with residual errors seem near to zero of mean and uniform variance. The Q-Q plot in the **bottom left** indicates a slight left skew in the distribution, suggesting a reasonable adherence.

The plot in the **bottom right** presents the correlogram, showing that the residual errors are not autocorrelated, indicating white noise.

### Step 6: Validate model

In this step, I will evaluate the ARIMA model using Root Mean Square Error (RMSE), Mean Squared Error (MSE), and R<sup>2</sup> score  to verify how well the model is able to predict future values based on the patterns learned from the training set. But first, I will use the 'forecast()' function to make predictions on the test data, based on the ARIMA model trained on the training dat (Brownlee, 2023)..

```python
for t in range(len(test_data)):
 model = ARIMA(train_ar, order=(0,1,0), missing='drop')
 model_fit = model.fit()
 output = model_fit.forecast()
 yhat = output[0]
 predictions.append(yhat)
 obs = test_data[t]
 train_ar.append(obs)
 print('predicted=%f, expected=%f' % (yhat, obs))
```

The metrics were computed below and resulted in a very good perfomance of the ARIMA model with 95% accuracy as R<sup>2</sup> score is 0.95 and low RMSE and MSE. The RMSE is directly interpretable as 0.41 cents considering that the data is about stock prices, mining low variation between actual values (test set) and the values predicted.

```python
rmse = sqrt(mean_squared_error(test_data, predictions))
print('Root Mean Square Error (RMSE): %.3f' % rmse)

mse = mean_squared_error(test_data, predictions)
print('Mean Squared Error (MSE): %.3f' % mse)

r_squared = r2_score(test_data, predictions)
print('R-squared score: %.3f' % r_squared)
```

The metrics resulted in an excellent model performance in which it can explain 95.5% of variance in the data, which is quite high. The Root Mean Square Error (RMSE) and Mean Squared Error (MSE) are very close to zero, indicating a good predictability of the model since the variation between actual and predicted values is close. 

In this visualization is showing the small difference between actual and predicted values.

```python
pyplot.plot(test_data, color='blue', label='Actual')
pyplot.plot(predictions, color='red', label='Predicted')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Actual vs. Predicted Values')
plt.legend()
pyplot.show()
```

# Forecast

### Using entire data (cour_filt['Close']):

In this step I will fit the ARIMA model (0,1,0) with the actual data (cour_filt['Close']). I use the '.asfreq()' method to add frequency information to forecasting. The letter 'B' represents 'Business Day', which is Monday through Friday, the day on which the stock market operates.

Comparing this AIC value with the one that used the 'train_data', we found an increase; however, the sigma2 coefficient was reduced, indicating a decrease of the noise term when using the entire data.  

```python
cour_filt = cour_filt.asfreq('B')

model_all_ts = ARIMA(cour_filt['Close'], order=(0,1,0))
fitted_model = model_all_ts.fit()
fitted_model.summary()
```

I will store the predicted values get from the ARIMA model, because it can be useful for evaluating the model performance on the entire data (cour_filt['Close']). Then I will add these values into time series (cour_filt).

```python
predictions = fitted_model.fittedvalues
cour_filt['ARIMA_pred'] = predictions
```

```python
predictions, cour_filt['Close']
```

Below, I will add the residual to the fitted values in order to obtain complete predicted values, that can better replect te observed data and improve the overall performance. Then I will add these values into time series (cour_filt).

```python
Res_Pred = fitted_model.resid + fitted_model.fittedvalues
cour_filt['Res_Pred'] = Res_Pred
```

```python
cour_filt.head()
```

Then, I will check and remove missing values from the time series (cour_filt).

```python
cour_filt.isnull().sum()
```

```python
cour_filt = cour_filt.dropna()
cour_filt.isnull().sum()
```

I will plot the actual values, which are the residuals added to the fitted values, and the forecast made by the ARIMA model (0,1,0).

```python
f, axarr = plt.subplots(1, sharex=True)
f.set_size_inches(10, 6)
cour_filt['Res_Pred'].iloc[1:].plot(color='g', linestyle='-', ax=axarr, label='Actual')
cour_filt['ARIMA_pred'].iloc[1:].plot(color='r', linestyle='--', ax=axarr, label='ARIMA Forecast')
axarr.set_title('ARIMA(0, 1, 0): COUR close price')
axarr.set_xlabel('Date')
axarr.set_ylabel('Close')
axarr.legend()
plt.show()
```

As we can see in this chart, the ARIMA forecast benefited greatly from the entire time series. However, metrics like mean_squared_error and R<sup>2</sup> score will be necessary to certify it.

```python
rmse = sqrt(mean_squared_error(cour_filt['Res_Pred'], cour_filt['ARIMA_pred']))
print('Root Mean Square Error (RMSE): %.3f' % rmse)

mse = mean_squared_error(cour_filt['Res_Pred'], cour_filt['ARIMA_pred'])
print('Mean Squared Error (MSE): %.3f' % mse)

r_squared = r2_score(cour_filt['Res_Pred'], cour_filt['ARIMA_pred'])
print('R-squared score: %.3f' % r_squared)
```

Considering the metrics, we can confirm a good model performance, with the model explaining 88% of the variance, which is considered high. RMSE and MSE are similar, with 1.20 and 1.44, respectively, but not satisfactory as the stock price varied by $1.20, which can be considered a significant variation in the case.

### Predicting future values:

To predict future values, I used all the data (cour_filt['Close']) to fit the ARIMA model. The first time I used these components (0,1,0), the predicted values were the same (all predicted values were equal). However, when I used the second-order differencing, the results seemed more compatible with the original data, which possibly helped remove any remaining trends and seasonality in the data, resulting in predictions that were more compatible with the original data.

I also experimented with other terms (p and q), but only with the second-order differencing did the model perform well.

```python
cour_filt = cour_filt.asfreq('B')

model_tsclose = ARIMA(cour_filt['Close'], order=(0,2,0))
model_tsclose_fitted = model_tsclose.fit()

model_tsclose_fitted.summary()
```

Below, I am using the steps equal to 5 to predict 5 values in the future with a 95% confidence interval (alpha=0.05). Then, I used the '.get_forecast()' method to generate the forecast based on the fitted model. Following, I used the 'summary_frame' to contain the predicted values with a 95% confidence level. 

As we can see, the predictions continue the index data, inclusive of the frequency in which the stock market operates, in accordance with the business day. 

```python
steps = 5
alpha = 0.05
forecast_10 = model_tsclose_fitted.get_forecast(steps=steps)
fc_and_interv_10 = forecast_10.summary_frame(alpha=alpha)
fc_and_interv_10
```

In the chart below, we can see the mean of the predicted values (blue) with a slight variation and following a decrescent tendency that has been in the original data. Another interesting fact is that the confidence interval increases as we move forward from the original data, suggesting that predicting for an extended period is unreliable with this model's adjustments. To solve this, probably, more adjustments will be necessary to capture underlying patterns in the data that are becoming more complex (less predictable) as time progresses.

```python
plt.figure(figsize=(10, 5))
plt.plot(fc_and_interv_10.index, fc_and_interv_10['mean'], label='Mean Forecast', color='blue')
plt.fill_between(fc_and_interv_10.index, fc_and_interv_10['mean_ci_lower'], fc_and_interv_10['mean_ci_upper'], color='blue', alpha=0.1, label='95% Confidence Interval')
plt.xlabel('Date')
plt.ylabel('Close')
plt.title('Forecast with Confidence Intervals: COUR')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

This chart below shows an overall comparison between the original data and the predicted values. Although the prediction period is small, it seems to follow the original trend.

```python
mse = mean_squared_error(cour_filt['Close'][-steps:], fc_and_interv_10['mean'])
print('Mean Squared Error (MSE): %.3f' % mse)

rmse = np.sqrt(mse)
print('Root Mean Square Error (RMSE): %.3f' % rmse)

r2 = r2_score(cour_filt['Close'][-steps:], fc_and_interv_10['mean'])
print('R-squared score: %.3f' % r_squared)
```
